```python
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
lyrics = pd.read_csv('lyrics.csv')
lyrics1 = lyrics.query('genre != "Not Available"')
lyrics1 = lyrics1.dropna()

#I played around with the following code for the second part of this notebook but the first part was run with
#each n=6000 and all 3 of lyrics2,3,4 in lyrics0
lyrics2 = lyrics1[lyrics1["genre"] == "Rock"].sample(n=3000)
lyrics3 = lyrics1[lyrics1["genre"] == "Pop"].sample(n=200)
lyrics4 = lyrics1[lyrics1["genre"] == "Hip-Hop"].sample(n=3000)
lyrics0 = pd.concat([lyrics2, lyrics4])
lyrics0 = lyrics0.reset_index(drop=True) #reorders the index so that we can enumerate over the df
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test = train_test_split(lyrics0, train_size=.7)

x_train = x_train.reset_index(drop=True)
```

```python
countvec = CountVectorizer().fit_transform(x_train["lyrics"])
tfidf  = TfidfTransformer().fit_transform(countvec)
```
